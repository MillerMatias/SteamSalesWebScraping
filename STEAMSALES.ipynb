```python
import bs4 #beautifulsoup4
from urllib.request import urlopen as uReq #URL-open modulen som vi importerar
from bs4 import BeautifulSoup as soup #beautifulsoup modulen som importeras 

myurl = "https://store.steampowered.com/search/?specials=1" #Steam-sale URLen som vi använde

uClient = uReq(myurl) #Laddar ner sidan som en client
page_html = uClient.read() # överför HTML till page_html svariabeln

uClient.close() #Stäng klienten

page_soup = soup(page_html, "html.parser") #Parsa själva HTML med beautifulsoup (funkar också med XML)


```

```python
containers = page_soup.findAll("div", {"class":"responsive_search_name_combined"}) #Visa all content inne i responsive_search_name_combined diven
containers
```

## Spelets titel

```python
#Ta fram spelets TITLE 
for container in containers: #Loopa igenom hela container variabeln som har egentligen hela sidan i sig
    title_container = container.find("span", ["class", "title"]) #Plocka fram span - class - title
    title = title_container.get_text() #Ta bort <span> tags med .get_text()
    print('\x1b[1;34m'+'Title: '+'\x1b[0m' + title) #Printa
    
    




```

## Plocka ut Rating (%) och Review (Mängden användare som har givit en rating) 

```python
#RATING & REVIEW
try:
    for container in containers: #Loopa igenom hela sidan igen
        findrating = container.find("div", ["class", "col search_reviewscore responsive_secondrow", "data-tooltip-html"]) #Gå in på rätt div som innehåller både rating(%) och antalet reviews
        ratingstring = findrating.find("span", ["class", "search_review_summary"]) #Gå ännu djupare för att få bara fram span-class-search-review_summary
        if ratingstring == None: #Om spelet inte har en rating så ska "None" ersättas med No Rating
            print ("No rating")
        else:
            rating = ratingstring.get("data-tooltip-html").split('%')[0] #Ta fram hela stringen som finns före "%" tecknet
            rating_result = ''.join([i for i in rating if i.isdigit()]) #Ränsa ut allt annat utom siffrorna före % tecknet.
            
            reviews = ratingstring.get("data-tooltip-html").split('%')[-1] #Gör samma sak, men ta string efter "%" teckne.
            review_amount = ''.join([i for i in reviews if i.isdigit()]) #Ränsa igen bort allt annat utom siffrorna ur stringen.
            print("Postive likes: ", rating_result, "%") #Printa procent ratingen.
            print("Number of reviews: ", review_amount) #Printa antalet folk som har ratet.
        
except AttributeError: #Måst ha en try - catch annars gnäller den.
    print()
```

## Ta ut rabatt procenten

```python
#DISCOUNT PERCENT
try:
    for container in containers: #Loopa igenom igen containern som innehåller allt.. we know the drill
        discountpercent = container.find("div", ["class", "col search_discount responsive_secondrow"]) #Ta fram den specifika divven som har procenten i sig.
        discount = discountpercent.find("span") #Ta fram endast discounted procenten.
        
        print(discount.get_text()) #Printa.
        
except AttributeError:
    print()
```

## Ta ut originella priset

```python
#Original price
try:
    for container in containers: 
            findprice = container.find("div", ["class", "col search_price discounted responsive_secondrow"]) #Samma so i discounted percent men vi går in i search_price.
            price = findprice.find("strike") #Men här tar vi originella priset som finns i strike- taggen.
        
            print(price.get_text()) #O sen printar vi igen.
             
except AttributeError:
    print()
```

## Rabattpriset

```python
#Discounted price

try:
    for container in containers:
            findprice = container.find("div", ["class", "col search_price discounted responsive_secondrow"]) #Samma div/klass som i originella priset.
            discountprice = container.find("div", ["class", "col search_price discounted responsive_secondrow"]).contents[-1].strip() #Använd strip för att få ut discount priset, för att originella priset finns också i samma string..
            
            print(discountprice)
            
            
except AttributeError:
    print()
```

## Utgivningsår

```python
#Release year
try:
    for container in containers:
            releaseyear = container.find("div", ["class", "col search_released responsive_secondrow"]).get_text() #search_released innehåller året.
            
            if(releaseyear == ""): #Några spel har inga utgivningsår, så om det inte finns ett årtal så blir det unknown
                releaseyear = "unknown"
            
            print(releaseyear)
except AttributeError:
    print()
```

## Operativsystemet som märks med en 1 om spelet är kompatibelt, 0 om inte

```python
#Operativesystem
try:
    for container in containers:
            OS = container.find("div", ["class", "col search_name ellipsis"]) ##Hitta var operativsystemet finns.
            OScontainer = OS.find("p") #Ta fram allt inne i <p> taggen.
            win = OScontainer.find("span", ["class", "platform_img win"]) #Plocka ut varsit operativsystem i en variabel.
            mac = OScontainer.find("span", ["class", "platform_img mac"])#Plocka ut varsit operativsystem i en variabel.
            lin = OScontainer.find("span", ["class", "platform_img linux"])#Plocka ut varsit operativsystem i en variabel.
            
            for span in OScontainer: #Loopa igenom <p> taggen som innehåller alla operativsystem.
                if (lin):            #Om det finns win/mac/lin så sätt en 1 till dom, annors blir det 0.
                    lin = 1
                    print (lin)
                else: print (0)
                    
                if (mac):
                    mac = 1
                    print (mac)
                else: print (0)
                    
                if (win):
                    win = 1
                    print (win)
                else: print (0)
                        
except AttributeError:
    print()
    

```

# Koden som helhet, kör denhär varje gång du vill ha en uppdaterad .CSV fil

```python
#All kod ihopsatt

import time #import time för timestampen
import bs4 #beautifulsoup4
from urllib.request import urlopen as uReq #URL-open modulen som vi importerar
from bs4 import BeautifulSoup as soup #beautifulsoup modulen som importeras 

myurl = "https://store.steampowered.com/search/?specials=1" #Steam-sale URLen som vi använde

uClient = uReq(myurl) #Laddar ner sidan som en client
page_html = uClient.read() # överför HTML till page_html variabeln

uClient.close() #Stäng klienten

page_soup = soup(page_html, "html.parser") #Parsa själva HTML med beautifulsoup (funkar också med XML)


timenow = time.strftime('%d-%m-%Y %H:%M:%S') #timenow variablen innehåller timestampen när koden körs.

containers = page_soup.findAll("div", {"class":"responsive_search_name_combined"}) #Visa all content inne i responsive_search_name_combined diven

try:
    filename = "steamsales.csv" #namn på filen
    f = open(filename, "w") #open file i skriv-mode

    headers = "Spelnamn|Rating|#Reviews|Rabatt%|Pris|OrdinariePris|Utgivningsar|Win|Lin|OSX|Tid\n" #första raden i excel filen (.csv)
    f.write(headers) # skriv headers till excel filen
    
    #LOOP CONTAINER
    for container in containers: 
        
        #TITLE
        title = container.find("span", ["class", "title"]).get_text()
    
        #DISCOUNTPERCENT
        discountpercent = container.find("div", ["class", "col search_discount responsive_secondrow"])
        discount = discountpercent.find("span").get_text()
            
        #DISCOUNTPRICE
        findprice = container.find("div", ["class", "col search_price discounted responsive_secondrow"])
        discountprice = container.find("div", ["class", "col search_price discounted responsive_secondrow"]).contents[-1].strip()
        
        #ORIGINALPRICE
        findprice = container.find("div", ["class", "col search_price discounted responsive_secondrow"])
        price = findprice.find("strike").get_text()
        
        #RELEASEYEAR
        releaseyear = container.find("div", ["class", "col search_released responsive_secondrow"]).get_text()
        
        if(releaseyear == ""):
                releaseyear = "unknown"

        #OPERATIVESYSTEM
        OS = container.find("div", ["class", "col search_name ellipsis"])
        OScontainer = OS.find("p")
        win = OScontainer.find("span", ["class", "platform_img win"])
        mac = OScontainer.find("span", ["class", "platform_img mac"])
        lin = OScontainer.find("span", ["class", "platform_img linux"])
        for span in OScontainer:
            if (lin):
                lin = int(1)
                #print (lin)
            else: lin = int(0)
                    
            if (mac):
                mac = int(1)
                #print (mac)
            else: mac = int(0)
                
            if (win):
                win = int(1)
                #print (win)
            else: win = int(0)   

        #RATING% AND #REVIEW
        findrating = container.find("div", ["class", "col search_reviewscore responsive_secondrow", "data-tooltip-html"])
        ratingstring = findrating.find("span", ["class", "search_review_summary"])
        if ratingstring == None:
            rating_result = "No rating"
            review_amount = "N/A"
        else:
            rating = ratingstring.get("data-tooltip-html").split('%')[0]
            rating_result = ''.join([i for i in rating if i.isdigit()])
          
            reviews = ratingstring.get("data-tooltip-html").split('%')[-1]
            review_amount = ''.join([i for i in reviews if i.isdigit()])
            
 
        #Skriv till CSV filen med | som delimiter..
        f.write(title + "|" + rating_result + "|" + review_amount + "|" + discount + "|" + discountprice + "|" + price + "|" + releaseyear + "|" + str(win) + "|" + str(lin) + "|" + str(mac) + "|" + timenow + "\n")
except AttributeError:
    print()
```
